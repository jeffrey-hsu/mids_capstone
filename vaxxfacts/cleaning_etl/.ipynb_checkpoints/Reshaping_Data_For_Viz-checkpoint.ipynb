```python
import numpy as np
import pandas as pd
```

```python
# Import compiled data
data1112 = pd.read_csv("cleanjoin1112.csv", encoding = "ISO-8859-1")
data1213 = pd.read_csv("cleanjoin1213.csv", encoding = "ISO-8859-1")
data1314 = pd.read_csv("cleanjoin1314.csv", encoding = "ISO-8859-1")
data1415 = pd.read_csv("cleanjoin1415.csv", encoding = "ISO-8859-1")
data1516 = pd.read_csv("cleanjoin1516.csv", encoding = "ISO-8859-1")
data1617 = pd.read_csv("cleanjoin1617.csv", encoding = "ISO-8859-1")
data1718 = pd.read_csv("cleanjoin1718.csv", encoding = "ISO-8859-1")
```

```python
#Keep county, city, school grade, population #, enrollment, vaccinated (n), 
vac1112=data1112[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1213=data1213[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1314=data1314[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1415=data1415[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1516=data1516[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1617=data1617[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
vac1718=data1718[['county','city','enrollment','n','vac_info_type', 'tot_pop']]
```

```python
vac1112.head()
```

```python
data1112.head()
```

```python
vac1112=vac1112.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1213=vac1213.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1314=vac1314.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1415=vac1415.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1516=vac1516.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1617=vac1617.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
vac1718=vac1718.groupby(by=['county','city','vac_info_type', 'tot_pop'], as_index=False)['enrollment','n'].sum()
```

```python
vac1112['year']='2011-2012'
vac1213['year']='2012-2013'
vac1314['year']='2013-2014'
vac1415['year']='2014-2015'
vac1516['year']='2015-2016'
vac1617['year']='2016-2017'
vac1718['year']='2017-2018'
```

```python
vac1112.head()
```

```python
vac_11_18=pd.concat([vac1112, 
           vac1213,
           vac1314,
           vac1415,
           vac1516,
           vac1617,
           vac1718
          ],ignore_index=True)

```

```python
vac_11_18.rename(columns={'n': 'vaccinated'}, inplace=True)
```

```python
vac_11_18.head()
```

```python
#Export full cleaned data
vac_11_18.to_csv("vac_11_18_combined.csv", index=False)
```
