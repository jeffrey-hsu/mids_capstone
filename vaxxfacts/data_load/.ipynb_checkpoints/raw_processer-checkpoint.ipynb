```python
import xlrd, csv
import itertools
import re
from operator import itemgetter
import os, json, getpass, imp
import processing_helper as h
import pandas as pd

BASE_PATH = "/Users/" + getpass.getuser() + "/capstone_data/raw_data/school_vaccination/"
```

**XLSX/XLS raw files to CSV**

```python
## Define the columns & the source raw xls/xlsx files structure

col_names = ['facility_num', 'county', 'is_public', 'city', 'facility_name',
             'enrollment', 'n', 'pct', 'reported', 'vac_info_type']
data_col_dict = json.load(open('data_struc_def.txt', 'r'))
data_col_dict
```

```python
## The raw files to process

DATA_SOURCE = "data_source.txt"
source_paths = json.load(open(DATA_SOURCE))
folders = list(source_paths['vaccination'].keys())
doc_dict = {} 
for f in folders:
    print("=====", f, "=====")
    docs = source_paths['vaccination'][f]
    fnames = []
    for doc in docs:
        fname = doc.split("/")[-1]
        fnames.append(fname)
        print(fname)
    doc_dict[f] = fnames
```

```python
## Process raw files and convert to csv
## with the predefined columns

imp.reload(h)
for k, vs in doc_dict.items():
    print("=====", k, "=====")
    for v in vs:
        print("file:", v)
        input_file_path = BASE_PATH + k + "/" + v
        output_file_name = "".join(v.split(".")[:-1]) + ".csv"
        output_file_path = BASE_PATH + k + "/" + output_file_name
        sh = h.xlsx2csv(input_file_path = input_file_path, 
                        output_file_path = output_file_path, 
                        col_dict = data_col_dict)
```

```python
## Look at an example file to check processing result

df = pd.read_csv(BASE_PATH + "/2017-2018/2017-18CA7thGradeData.csv"
                 , skipinitialspace = True, sep = ",", names = col_names)
df.shape
```

**Merge CSV files**

```python
## Define the csv files

csv_files = {}
for i, folder in enumerate(source_paths['vaccination']):
    #print(i, folder)
    file_list = []
    for path in source_paths['vaccination'][folder]:
        file = path.split('/')[-1].split('.')[0] + '.csv'
        file_list.append(file)
    csv_files[folder] = file_list
```
